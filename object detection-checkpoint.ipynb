```python
# Real time face detection project
import cv2
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
webcam = cv2.VideoCapture(0)
while True:
    _,img= webcam.read()
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray,1.5,4)
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
    cv2.imshow("face detection",img)
    key = cv2.waitKey(10)
    if key == 27:
        break
webcam.release()
cv2.destroyAllWindows()
```

```python
#simple webcam starting program using opencv
import cv2

# Initialize the webcam (camera index 0)
cam = cv2.VideoCapture(0)

# Create a named window for display
cv2.namedWindow("Webcam Feed")

# Counter for image filenames
img_counter = 0

while True:
    # Read a frame from the webcam
    ret, frame = cam.read()
    if not ret:
        print("Failed to grab frame")
        break

    # Display the frame
    cv2.imshow("Webcam Feed", frame)

    # Wait for a key press
    k = cv2.waitKey(1)

    if k == 27:  # ESC key pressed
        print("Escape hit, closing...")
        break
    elif k == 32:  # SPACE key pressed
        img_name = f"opencv_frame_{img_counter}.png"
        cv2.imwrite(img_name, frame)
        print(f"{img_name} written!")
        img_counter += 1

# Release the webcam and close the window
cam.release()
cv2.destroyAllWindows()
```

```python
#object detection and size measurement 
import cv2

# Load the image from the webcam
imgs = cv2.VideoCapture(0)
ret, frame = imgs.read()

if ret:
    # Save the captured frame
    cv2.imwrite('capture.jpg', frame)
    print("Image saved successfully!")

    # Display the captured frame
    cv2.imshow("Webcam Feed", frame)
    cv2.waitKey(0)  # Wait for a key press to close the window

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Threshold the grayscale image
    ret, thresh = cv2.threshold(gray, 127, 255, 0)

    # Find contours in the thresholded image
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    # Draw the contours on the original frame
    cv2.drawContours(frame, contours, -1, (0, 255, 0), 3)

    # Get the area of the object in pixels
    area = cv2.contourArea(contours[0])

    # Convert the area from pixels to a real-world unit of measurement (e.g., cm^2)
    scale_factor = 0.1  # 1 pixel = 0.1 cm
    size = area * scale_factor ** 2

    # Print the size of the object
    print('Size:', size)

    # Draw a bounding box around the detected object
    x, y, w, h = cv2.boundingRect(contours[0])
    cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

    # Save the image with the contours and bounding box
    cv2.imwrite('object_with_contours.jpg', frame)
else:
    print("Failed to grab frame from the camera.")
```
